```r
# True value 
# need true X and true Y
# need epsilon and delta
# then we can get (X, Y)
# need to generate Z and err
generate.data = function(N, beta0, beta1){
  epsilon = rnorm(N)
  delta = rnorm(N)
  err = rnorm(N)
  X.star = rnorm(N, 0, 2)
  Y.star = beta0 + beta1 * X.star
  Y = Y.star + epsilon
  X = X.star + delta
  Z1 = X.star + err
  Z2 = 0.1 * X.star + err
  data = data.frame(Y_star = Y.star, X_star = X.star, Y = Y, X = X, Z1 = Z1, Z2 = Z2)
  return (data)
}
```

```r
dt = generate.data(N = 300, beta0 = -1, beta1 = 1)
print(head(dt))
```

```r
beta = c()
for (i in 1:1000){
    dt = generate.data(N = 300, beta0 = -1, beta1 = 1)
    x = dt[, 4]
    y = dt[, 3]
    z1 = dt[, 5]
    z2 = dt[, 6]
    beta1.hat.LS = sum((x - mean(x)) * (y - mean(y))) / sum((x - mean(x))^2)
    beta2.M = sum((x - mean(x)) * (y - mean(y))^2) / sum((x - mean(x))^2 * (y - mean(y)))
    beta3.IV1 = sum((z1 - mean(z1)) * (y - mean(y))) / sum((z1 - mean(z1)) * (x - mean(x)))
    beta3.IV2 = sum((z2 - mean(z2)) * (y - mean(y))) / sum((z2 - mean(z2)) * (x - mean(x)))
    tmp = c(beta1.hat.LS, beta2.M, beta3.IV1, beta3.IV2)
    beta = rbind(beta, tmp)
}
```

```r
beta1.set = c()
se_beta1.set = c()
cp.set = c()
beta1.boot.set = c()
for (i in 1:1000){
    dat = generate.data(N = 300, beta0 = -1, beta1 = 1)
    x = dat[, 4]
    y = dat[, 3]
    fit = lm(y ~ x)
    coef = summary(fit)$coef[c(1, 2)]
    gam0 = summary(fit)$coef[1]
    gam1 = summary(fit)$coef[2]
    Sxx = mean((x - mean(x))^2)
    sigma2_x = Sxx - 1
    k = sigma2_x / Sxx
    beta1 = gam1 / k
    beta0 = mean(y) - beta1 * mean(x)
    se_gam1 = summary(fit)$coef[2, 2]
    # b1 is true value
    b1 = 1
    se_beta1 = se_gam1 / k
    z_test = (beta1 - b1) / se_beta1
    cp = 1 * (abs(z_test) <= 1.96)
    beta1.set = c(beta1.set, beta1)
    se_beta1.set = c(se_beta1.set, se_beta1)
    cp.set = c(cp.set, cp)

    
    beta1_boot_set = c()
    for (j in 1:80){
        index = sample(c(1:300), 300, replace = T)
        data_boot = dat[index, ]
        x_boot = data_boot[, 4]
        y_boot = data_boot[, 3]
        fit_boot = lm(y_boot~x_boot)
        coef_boot = summary(fit_boot)$coef[c(1, 2)]
        gam0_boot = summary(fit_boot)$coef[1]
        gam1_boot = summary(fit_boot)$coef[2]
        Sxx_boot = mean((x_boot - mean(x_boot))^2)
        sigma2_x_boot = Sxx_boot - 1
        k_boot = sigma2_x_boot / Sxx_boot
        beta1_boot = gam1_boot / k_boot
        beta0_boot = mean(y_boot) - beta1_boot * mean(x_boot)
        se_gam1_boot = summary(fit_boot)$coef[2, 2]
        se_beta1_boot = se_gam1_boot / k_boot
        z_test_boot = (beta1_boot - b1) / se_beta1_boot
        cp_boot = 1 * (abs(z_test_boot) <= 1.96)
        beta1_boot_set = c(beta1_boot_set, beta1_boot)
    }
    beta1.boot.set = c(beta1.boot.set, mean(beta1_boot_set))
}
```
